```python
#var 6 task 2
import numpy as np

def process_matrix(input_file, output_file):
    # Считываем матрицу из файла
    with open(input_file, 'r') as f:
        N = int(f.readline().strip())  # Читаем размерность матрицы
        matrix = []
        for _ in range(N):
            row = list(map(int, f.readline().strip().split()))
            matrix.append(row)
    
    # Преобразуем в NumPy массив
    matrix = np.array(matrix)

    # Находим наибольший элемент в главной и побочной диагоналях
    main_diagonal = [matrix[i][i] for i in range(N)]
    secondary_diagonal = [matrix[i][N - i - 1] for i in range(N)]
    diagonal_elements = main_diagonal + secondary_diagonal
    max_element = max(diagonal_elements)
    
    # Находим индекс максимального элемента
    if max_element in main_diagonal:
        max_index = main_diagonal.index(max_element)
    else:
        max_index = N - 1 - secondary_diagonal.index(max_element)

    # Элемент на пересечении диагоналей
    center_index = N // 2
    center_element = matrix[center_index][center_index]

    # Меняем местами
    matrix[max_index][max_index] = center_element
    matrix[center_index][center_index] = max_element

    # Записываем результаты в файл
    with open(output_file, 'w') as f:
        f.write("Матрица после изменения:\n")
        for row in matrix:
            f.write(' '.join(map(str, row)) + '\n')
        f.write(f"Наибольший элемент: {max_element}\n")

# Пример использования
process_matrix('Шураев_УБ41_vvod.txt', 'Шураев_УБ41_vivod.txt')
```
